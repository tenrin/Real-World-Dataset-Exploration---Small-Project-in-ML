```python
subset_variables = ['congestion', 'high_fever', 'cough', 'runny_nose', 'chills']
corr = train_df[subset_variables].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Split the columns into four groups
num_columns = len(train_df.columns)
quarter_num_columns = num_columns // 4

first_group_columns = train_df.columns[:quarter_num_columns]
second_group_columns = train_df.columns[quarter_num_columns:2*quarter_num_columns]
third_group_columns = train_df.columns[2*quarter_num_columns:3*quarter_num_columns]
fourth_group_columns = train_df.columns[3*quarter_num_columns:]

# Create separate plots for the four groups of columns
plt.figure(figsize=(20, 15))

# Plot box plot for the first group of columns
plt.subplot(2, 2, 1)
train_df[first_group_columns].boxplot(figsize=(15, 6))
plt.title('Box Plot for First Group of Columns')
plt.xticks(rotation=90)

# Plot box plot for the second group of columns
plt.subplot(2, 2, 2)
train_df[second_group_columns].boxplot(figsize=(15, 6))
plt.title('Box Plot for Second Group of Columns')
plt.xticks(rotation=90)

# Plot box plot for the third group of columns
plt.subplot(2, 2, 3)
train_df[third_group_columns].boxplot(figsize=(15, 6))
plt.title('Box Plot for Third Group of Columns')
plt.xticks(rotation=90)

# Plot box plot for the fourth group of columns
plt.subplot(2, 2, 4)
train_df[fourth_group_columns].boxplot(figsize=(15, 6))
plt.title('Box Plot for Fourth Group of Columns')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
symptoms = ['congestion', 'high_fever', 'cough', 'runny_nose', 'chills']
correlation_matrix = train_df[symptoms].corr()
print(correlation_matrix)

plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
Y_train = pd.DataFrame(train_df['prognosis'])
X_train = train_df.drop('prognosis', axis = 1)
```

```python
Y_train
```

```python
from scipy.stats import chi2_contingency

symptoms_to_analyze = ['runny_nose', 'congestion', 'chills', 'high_fever']

# Calculate chi-square statistics for each symptom
chi2_values = []
p_values = []
for feature in symptoms_to_analyze:
    contingency_table = pd.crosstab(X_train[feature], Y_train['prognosis'])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    chi2_values.append(chi2)
    p_values.append(p) 
    print(f"Feature: {feature}")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print("\n")
```

```python
symptoms_to_analyze = ['vomiting', 'fatigue', 'high_fever']
grouped = train_df.groupby('prognosis')[symptoms_to_analyze].std()
print("Variability of Symptoms Across Diseases:")
print(grouped)

grouped.boxplot(figsize=(10, 6))
plt.title('Variability of Symptoms Across Diseases')
plt.ylabel('Standard Deviation')
plt.xlabel('Symptoms')
plt.xticks(rotation=45)
plt.show()
```

```python
symptom_occurrences = X_train.sum()
low_threshold = 350
high_threshold = 750

# Splitting symptom occurrences into 3 grps
top_low_occurrence_symptoms = symptom_occurrences[symptom_occurrences < low_threshold]
top_medium_occurrence_symptoms = symptom_occurrences[(symptom_occurrences >= low_threshold) & (symptom_occurrences <= high_threshold)]
top_occurrence_symptoms = symptom_occurrences[symptom_occurrences > high_threshold]
```

```python
 #Plotting top occurrence symptoms
plt.figure(figsize=(10, 6))
top_occurrence_symptoms.sort_values().plot(kind='barh', color='SKYBLUE')
plt.title('Symptoms with high occurence')
plt.xlabel('Count')
plt.ylabel('Symptoms')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_medium_occurrence_symptoms = top_medium_occurrence_symptoms.sort_values(ascending=False).head(15)
top_medium_occurrence_symptoms.sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 15 Occurrence Symptoms from Medium Occurrence Group')
plt.xlabel('Count')
plt.ylabel('Symptoms')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_low_occurrence_symptoms = top_low_occurrence_symptoms.sort_values(ascending=False).head(15)
top_low_occurrence_symptoms.sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 15 Occurrence Symptoms from Low Occurrence Group')
plt.xlabel('Count')
plt.ylabel('Symptoms')
plt.tight_layout()
plt.show()
```
