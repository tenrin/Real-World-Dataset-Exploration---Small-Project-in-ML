4. Drop the columns that have more than 90% missing values. 
        - define a list columns_to_drop containing the column names to drop 
        - Create a new dataframe ebola_droppedCol
        - display the shape of the new dataframe.
        - print the number of missing values in ebola_droppedCol
        

```python
missing_val_percent = (ebola.isnull().sum() / len(ebola)) * 100
columns_to_drop = missing_val_percent[missing_val_percent > 90].index.tolist()

ebola_droppedCol = ebola.drop(columns=columns_to_drop)
print("Shape of ebola_droppedCol:", ebola_droppedCol.shape)

print("Number of missing values in ebola_droppedCol:")
print(ebola_droppedCol.isnull().sum())
```

5. Answer each of the following questions: 
    - Count the number of missing values in the dataframe after the fill backward imputation of missing values. show the first 10 rows and 5 columns after imputation. 
    - Explain what changes have been made after fill backward imputation of the missing values and why. 


```python
ebola.iloc[0:10, 0:6]
```

```python
ebola_filled = ebola.fillna(method='bfill')

missing_values_after_imputation = ebola_filled.isnull().sum().sum()
print(f"Number of missing values after fill backward imputation: {missing_values_after_imputation}")

print("First 10 rows and 5 columns after fill backward imputation:")
ebola_filled.iloc[:10, :5]
```

- After the fill backward imputation is performed, the missing values in cell are filled using the next non-missing values in the same column. This method is widely practised in time series data.. But this method may not be suitable for all scenarios.

- After the fill forward imputation is performed, the missing values in cell are filled using the previous non-missing values in the same column. Just like fill backward imputation, this method is widely practised in time series data.

9. Write a paragraph to highlight the differences between the fill foward/backward versus linear interpolation of the missing values. 



- Fill Forward imputation fills the missing data using the data from previous row. Fill Backward imputation fill the missing data using the data from next row. Both of these methods are widely used in ML and Data Science where the data follows a time series order. On the other hand, Linear interpolation calculates the missing data based on the linear trend between the two adjacent known values. This method is a little more sophisticated and can be employed when the data does not follow a time series pattern.

10. Time series interpolation works by calling interpolate with the option method='time'.
    - Explain how is it different than the linear interpolation?

- Linear interpolation estimates a linear relationship between adjacent observation; for this we choose (method = 'linear') in the interpolate(). Time interpolation assume the time intervals between the rows. for this we choose (method = 'time') Time interpolation is only designed for time-series data.
